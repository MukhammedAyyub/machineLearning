```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
data = pd.read_csv('Salary_Data.csv')
df = pd.DataFrame(data)
df
```

```python
df = pd.DataFrame(data)
df
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = df['YearsExperience']
y = df['Salary']
x = np.array(x).reshape(-1, 1)
y = np.array(y).reshape(-1, 1)


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x = np.array(x).reshape(-1, 1)
y = np.array(y).reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("Test Data (X_test):", X_test.flatten())
print("Actual Cost (y_test):", y_test)
print("Predicted Cost (y_pred):", y_pred)
```

```python
plt.scatter(X_train, y_train, c='blue', label='Training Data')
plt.scatter(X_test, y_test, c='green', label='Test Data')
plt.plot(x, model.predict(x), c='red', ls='--', label='Regression Line')
plt.xlabel('Working experience (year)')
plt.ylabel('Amount of salary per month (USD)')
plt.title('Liner regression of salaries')
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```
